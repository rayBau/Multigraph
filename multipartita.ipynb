```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
G = nx.Graph()
```

```python
G.add_edge("s1","Bilbao", wight = 13)
G.add_edge("s1","Barcelona", wight = 13)
G.add_edge("s1","Sevilla", wight = 13)

G.add_edge("s2","Bilbao", wight = 13)
G.add_edge("s2","Barcelona", wight = 13)
G.add_edge("s2","Sevilla", wight = 13)
G.add_edge("s2","Valencia", wight = 13)
G.add_edge("s2","Zaragoza", wight = 13)

G.add_edge("s3","Sevilla", wight = 13)
G.add_edge("s3","Zaragoza", wight = 13)

```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
G = nx.Graph() 
```

```python
E = [("s1", "Bilbao", 4), ("s1", "Barcelona", 2), ("s1", "Sevilla", 2), 
     ("s2", "Bilbao", 1), ("s2", "Barcelona", 3), ("s2", "Sevilla", 2), ("s2", "Valencia",2),("s2", "Zaragoza", 3),
     ("s3", "Sevilla", 3), ("s3", "Zaragoza", 1)]
```

```python
G.add_weighted_edges_from(E)
```

```python
pos=nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight="bold")
edge_weight = nx.get_edge_attributes(G,"weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_weight)
plt.show()
```

```python
from networkx.algorithms import bipartite
```

```python
b = nx.Graph()
```

```python
b.add_nodes_from(["RH"], layer = 0)
b.add_nodes_from(["s1","s2","s3"], layer = 1)
b.add_nodes_from(["Bilbao", "Barcelona", "Sevilla", "Valencia", "Zaragoza"], layer = 2)
b.add_nodes_from(["FINAL"], layer = 3)
```

```python
E = [("RH", "s1", ""), ("RH", "s2", ""), ("RH", "s3", ""), 
     ("s1", "Bilbao", "5-3"), ("s1", "Barcelona", "------------2"), ("s1", "Sevilla", "------------2"), 
     ("s2", "Bilbao", 1), ("s2", "Barcelona", 3), ("s2", "Sevilla", 2), ("s2", "Valencia","------------2"),("s2", "Zaragoza", 3),
     ("s3", "Sevilla", "3"), ("s3", "Zaragoza", 1),
     ("Bilbao", "FINAL", ""),("Barcelona", "FINAL", ""),("Sevilla", "FINAL", ""),("Valencia", "FINAL", ""),("Zaragoza", "FINAL", "")]
```

```python
b.add_weighted_edges_from(E)
```

```python
edges = b.edges()
```

```python
pos = nx.multipartite_layout(b, subset_key="layer",)
plt.figure(figsize=(8, 5))
nx.draw(b, pos, with_labels=True, font_weight="bold")
edge_weight = nx.get_edge_attributes(b,"weight")
nx.draw_networkx_edge_labels(b, pos, edge_labels = edge_weight)
plt.savefig("multipartite_graph.png")
plt.show()
```
